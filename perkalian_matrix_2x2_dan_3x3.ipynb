```python
#Rumus
#input matriks 2x2 dan 3x3
def inputmatrix2x2():
    while True:
        try:
            namafile = input("Masukkan nama file matrix, file txt: ")
            with open(namafile, 'r') as file:
                matrix_strings = file.readlines()
                matrices = []
                temp_matrix = []
            for row in matrix_strings:
                if row.strip() == '---':
                    matrices.append(temp_matrix)
                    temp_matrix = []
                    continue
                numbers = row.strip().split(' ')
                numbers = [float(x) for x in numbers]
                temp_matrix.append(numbers)
            matrices.append(temp_matrix)
            print("===  Matrix yang terbaca dalam file ===")
            for i, matrix in enumerate(matrices):
                print("Matrix {}:".format(i+1))
                for row in matrix:
                    print(row)
                print()
            matrix1 = matrices[0]
            matrix2 = matrices[1]
            break
        except FileNotFoundError:
            print("File tidak dapat ditemukan. Masukkan nama file yang benar.")
    print("===  Bentuk persamaan linear matriks 2x2 ===")
    print()
    variables = ['x', 'y']
    for i, row in enumerate(matrix1):
        equation = []
        for j, item in enumerate(row):
            if item != 0:
                if item == 1:
                    equation.append(f"{variables[j]}")
                else:
                    equation.append(f"{item} {variables[j]}")
        if len(equation) > 0:
            print(f"{' + '.join(equation)} = {matrix2[i][0]}")

#rumus bentuk aljabar linear 3x3
def inputmatrix3x3():
    while True:
        try:
            namafile = input("Masukkan nama file matrix, file txt: ")
            with open(namafile, 'r') as file:
                matrix_strings = file.readlines()
                matrices = []
                temp_matrix = []
            for row in matrix_strings:
                if row.strip() == '---':
                    matrices.append(temp_matrix)
                    temp_matrix = []
                    continue
                numbers = row.strip().split(' ')
                numbers = [float(x) for x in numbers]
                temp_matrix.append(numbers)
            matrices.append(temp_matrix)
            print("===  Matrix yang terbaca dalam file ===")
            for i, matrix in enumerate(matrices):
                print("Matrix {}:".format(i+1))
                for row in matrix:
                    print(row)
                print()
            matrix1 = matrices[0]
            matrix2 = matrices[1]
            break
        except FileNotFoundError:
            print("File tidak dapat ditemukan. Masukkan nama file yang benar.")
    print("===  Variabel untuk persamaan linear matriks 3x3 ===")
    variables = [['x'],['y'],['z']]
    for i in variables:
        print(i)
    print()
    print("===  Bentuk persamaan linear matriks 3x3 ===")
    print()
    variables = ['x', 'y', 'z']
    for i, row in enumerate(matrix1):
        equation = []
        for j, item in enumerate(row):
            if item != 0:
                if item == 1:
                    equation.append(f"{variables[j]}")
                else:
                    equation.append(f"{item} {variables[j]}")
        if len(equation) > 0:
            print(f"{' + '.join(equation)} = {matrix2[i][0]}")


#Pilihan menu input
def menu():
    print('======== INPUT MATRIKS ========')
    print()
    print('Pilih ordo matriks:')
    print('1. Ordo 2x2')
    print('2. Ordo 3x3')
    print('3. Exit')
    print()
    pilihmenu()

#Pilih menu
def pilihmenu():
    pilih_menu = input('Pilih menu: ')
    if pilih_menu=='1':
        print('=== Input Matriks ordo 2x2 ===')
        print()
        inputmatrix2x2()
        hitunglagi()
    elif pilih_menu=='2':
        print('=== Input Matriks ordo 3x3 ===')
        print()
        inputmatrix3x3()
        hitunglagi()
    elif pilih_menu=='3':
        print('=== Keluar ===')
        exit()
    else:
        print('Pilihan menu salah')
        print()

def hitunglagi():
    pilih = input('Kembali ke menu? (y/t)')
    print()
    if pilih in ['y', 'Y', 'yes', 'Yes', 'YES', 'ya', 'YA','Ya']:
        menu()
    else:
        exit("Anda Keluar")
```
